#Feature engineering

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

# Binary columns
binary_cols = ['Gender','Married','Self_Employed','Education', 'Credit_History']

# Ordinal column
ordinal_cols = ['Dependents', 'Loan_Status']

# Nominal column
nominal_cols = ['Property_Area']

numeric_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']

# Defining transformations for binary, ordinal, and nominal, numerical variables
binary_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('binary', OrdinalEncoder())  # Treat binary columns as 0 and 1
])

ordinal_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ordinal', OrdinalEncoder(categories=[
        ['0', '1', '2', '3+'],   # Dependents categories
        ['N', 'Y']              # Loan_Status categories
    ]))
])

nominal_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(drop='first'))  # One-Hot Encode 'Property_Area'
])

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())  # Scale numeric variables
])

# Combining them into ColumnTransformer
preprocessor = ColumnTransformer(transformers=[
    ('binary', binary_transformer, binary_cols),
    ('ordinal', ordinal_transformer, ordinal_cols),
    ('nominal', nominal_transformer, nominal_cols),
    ('numeric', numeric_transformer, numeric_cols)
])

#Applying the transformation to the dataset
transformed_loan = preprocessor.fit_transform(loan)
```

```python
# Getting the column names after transformation
binary_cols_transformed = binary_cols
ordinal_cols_transformed = ordinal_cols
nominal_cols_transformed = preprocessor.named_transformers_['nominal']['onehot'].get_feature_names_out(nominal_cols)
numeric_cols_transformed = numeric_cols

# Combining all column names
new_column_names = list(binary_cols_transformed) + list(ordinal_cols_transformed) + list(nominal_cols_transformed) + list(numeric_cols_transformed)

# Converting the numpy array back to a DataFrame with appropriate column names
transformed_loan_df = pd.DataFrame(transformed_loan, columns=new_column_names)

print(transformed_loan_df.head())
```

#Neural Network

```python
# Defining a neural network with batch normalization and dropout
class LoanNN(nn.Module):
    def __init__(self):
        super(LoanNN, self).__init__()
        self.fc1 = nn.Linear(X.shape[1], 32)
        self.bn1 = nn.BatchNorm1d(32)
        self.dropout1 = nn.Dropout(0.3)  # Dropout layer after first batch normalization
        self.fc2 = nn.Linear(32, 16)
        self.bn2 = nn.BatchNorm1d(16)
        self.dropout2 = nn.Dropout(0.3)
        self.fc3 = nn.Linear(16, 1)

    def forward(self, x):
        x = self.fc1(x)
        x = self.bn1(x)
        x = torch.relu(x)
        x = self.dropout1(x)
        x = self.fc2(x)
        x= self.bn2(x)
        x = torch.relu(x)
        x = self.dropout2(x)
        x = self.fc3(x)
        return x
```

##Model training

```python
# Initializing the model, loss function, and optimizer
model2 = LoanNN()
random_state=42
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model2.parameters(), lr=0.0005)
```

```python
def accuracy_fn(y_true, y_pred):
    correct = torch.eq(y_true, y_pred).sum().item()
    acc = (correct / len(y_pred)) * 100
    return acc
```

```python
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32)
# Training loop
epochs = 400
for epoch in range(epochs):
    model2.train()  # Setting the model to training mode
    y_logits = model2(X_train_tensor).squeeze()  # Forward pas
    y_train_sq = y_train_tensor.squeeze()
    y_pred = torch.round(torch.sigmoid(y_logits)).squeeze()

    loss = criterion(y_logits, y_train_sq)
    acc = accuracy_fn(y_true=y_train_sq, y_pred=y_pred)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()


    model2.eval()  # Setting the model to evaluation mode
    with torch.inference_mode():
      test_logits = model2(X_test_tensor)
      test_pred = torch.round(torch.sigmoid(test_logits))
      # Ensuring y_test is a 2D tensor of shape (77, 1)
      y_test_tensor = torch.tensor(y_test, dtype=torch.float32).unsqueeze(1)
      test_loss = criterion(test_logits, y_test_tensor)

      test_acc = accuracy_fn(y_true=y_test_tensor, y_pred=test_pred)
    if epoch % 10 == 0:
      print(f"Epoch: {epoch} | Loss: {loss:.4f} | Accuracy: {acc:.2f}% | Test Loss: {test_loss:.5f} | Test acc: {test_acc: .2f}" )
```

1. Balancing the classes for Neural Networks decreased the accuracy of the test set to 71%. Therefore I did not perform the balancing of classes by BCEWithLogitsLoss with the pos_weight paramet.

2. Decreasing learning rate from 0,001% to 0,0005% increased the test accuracy from 79% to 82%

3. Increasing number of layers as well as size of the layers did not increase the accuracy of the test set and created overfitting. Therefore the number of layers remains 2.
