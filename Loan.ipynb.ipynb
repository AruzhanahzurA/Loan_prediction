```python
loan['Gender']=loan['Gender'].replace({'Male':1, 'Female':0}).fillna(-1)
loan['Married']=loan['Married'].replace ({'Yes':1, 'No':0}).fillna(-1)
loan['Education']=loan['Education'].replace({'Graduate':1, 'Not Graduate':0}).fillna(-1)
loan['Self_Employed']=loan['Self_Employed'].replace({'Yes':1, 'No':0}).fillna(-1)
loan['Property_Area']=loan['Property_Area'].replace({'Rural':0, 'Semiurban':1, 'Urban':2}).fillna(-1)
loan['Loan_Status']=loan['Loan_Status'].replace({'N':0, 'Y':1}).fillna(-1)
loan['Dependents']=loan['Dependents'].replace({'0':0, '1':1, '2':2, '3+':3}).fillna(-1)
loan['Credit_History']=loan['Credit_History'].fillna(-1)
```

```python
loan['ApplicantIncome']=loan['ApplicantIncome'].fillna(loan['ApplicantIncome'].mean())
loan['CoapplicantIncome']=loan['CoapplicantIncome'].fillna(loan['CoapplicantIncome'].mean())
loan['LoanAmount']=loan['LoanAmount'].fillna(loan['LoanAmount'].mean())
loan['Loan_Amount_Term']=loan['Loan_Amount_Term'].fillna(loan['Loan_Amount_Term'].mean())
```

```python
from sklearn.preprocessing import StandardScaler

norm_scaler = StandardScaler()
loan['ApplicantIncome'] = norm_scaler.fit_transform(loan[['ApplicantIncome']])
loan['CoapplicantIncome'] = norm_scaler.fit_transform(loan[['CoapplicantIncome']])
loan['LoanAmount'] = norm_scaler.fit_transform(loan[['LoanAmount']])
loan['Loan_Amount_Term'] = norm_scaler.fit_transform(loan[['Loan_Amount_Term']])
loan
```

```python
loan1=loan.drop(['Unnamed: 0','Loan_ID'], axis=1)
```

```python
loan1.isna().sum()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np

X=loan1.drop(columns=['Loan_Status'])
y=loan1['Loan_Status']

X_np_array=X.values
y_np_array=y.values

X_np_array = X_np_array.astype(np.float32)

X_tensor = torch.tensor(X_np_array, dtype=torch.float32)
y_tensor = torch.tensor(y_np_array, dtype=torch.float32).view(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(X_tensor, y_tensor, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(X.shape[1], 3)  # Input layer to hidden layer
        self.fc2 = nn.Linear(3, 2)
        self.fc3 = nn.Linear(2, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))  # Applying ReLU activation function
        x = torch.relu(self.fc2(x))  # Applying ReLU activation function
        x = self.fc3(x)
        return x
```

```python

# Initialize the model, loss function, and optimizer
model = SimpleNN()
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python

# Training loop
epochs = 300
for epoch in range(epochs):
    optimizer.zero_grad()  # Clear gradients
    outputs = model(X_train)  # Forward pass
    loss = criterion(outputs, y_train)  # Compute loss
    loss.backward()  # Backward pass
    optimizer.step()  # Update weights
    print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')
```

```python
# Evaluate the model on the test set
model.eval()  # Set the model to evaluation mode
with torch.no_grad():
    predictions = model(X_test)
    test_loss = criterion(predictions, y_test)
print(f'Test Loss: {test_loss.item():.4f}')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

linreg = LinearRegression()
linreg.fit(X_np_array, y_np_array)

predicted_heights = linreg.predict(X_np_array)
mse = mean_squared_error(y_np_array, predicted_heights)

# Print the mean squared error
print(mse)
```

Let`s add batch normalization and Dropouts

```python
# Define a neural network with batch normalization and dropout
class LoanNN(nn.Module):
    def __init__(self):
        super(LoanNN, self).__init__()
        self.fc1 = nn.Linear(X.shape[1], 64)
        self.bn1 = nn.BatchNorm1d(64)
        self.dropout1 = nn.Dropout(0.05)  # Dropout layer after first batch normalization
        self.fc2 = nn.Linear(64, 64)
        self.bn2 = nn.BatchNorm1d(64)
        self.dropout2 = nn.Dropout(0.05)  # Dropout layer after second batch normalization
        self.fc3 = nn.Linear(64, 64)
        self.bn3 = nn.BatchNorm1d(64)
        self.dropout3 = nn.Dropout(0.05)  # Dropout layer after third batch normalization
        self.fc4 = nn.Linear(64, 1)

    def forward(self, x):
        x = self.fc1(x)
        x = self.bn1(x)
        x = torch.relu(x)
        x = self.dropout1(x)
        x = self.fc2(x)
        x = self.bn2(x)
        x = torch.relu(x)
        x = self.dropout2(x)
        x = self.fc3(x)
        x = self.bn3(x)
        x = torch.relu(x)
        x = self.dropout3(x)
        x = self.fc4(x)
        return x
```

```python

# Initialize the model, loss function, and optimizer
model2 = LoanNN()
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python

# Training loop
epochs = 300
for epoch in range(epochs):
    optimizer.zero_grad()  # Clear gradients
    outputs = model2(X_train)  # Forward pass
    loss = criterion(outputs, y_train)  # Compute loss
    loss.backward()  # Backward pass
    optimizer.step()  # Update weights
    if (epoch+1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')
```
